```python
import matplotlib.pyplot as plt
import torch

mask = torch.triu(torch.ones(10, 10)).T
mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))

plt.imshow(mask)
plt.colorbar()
```
